```python
def plot_deg_dist(G):
    all_degs = list(dict(nx.degree(G)).values())
    unique_degs = list(set(all_degs))

    count_deg = []

    for i in unique_degs:
        x = all_degs.count(i)
        count_deg.append((x))
    plt.plot(unique_degs, count_deg, 'yo-')
    #plt.loglog(unique_degs, count_deg, 'yo-')
    plt.xlabel('Degrees')
    plt.ylabel('Number of nodes')
    plt.title('Degree distribution of karate network')
    plt.show()
```

```python
plot_deg_dist(G)
```

```python
H = nx.DiGraph()
for u,v in G.edges():
    if H.has_edge(u,v):
        H[u][v]['weight'] += 1
    else:
        H.add_edge(u, v, weight=1)

clustering = nx.clustering(H)
clustering
```
