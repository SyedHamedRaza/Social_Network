# Import Libraries

```python
import networkx
import networkx as nx
import matplotlib.pyplot as plt
```

# Graph APIs

```python
Z = nx.complete_graph(10)
print (Z.nodes()) # display all the nodes
print (Z.edges()) #print all the edges
print (Z.order()) #print number of nodes in a graph
print (Z.size()) #print number of edges in graph G
nx.draw(Z, with_label=1) #display graph Z
```

```python
H = nx.complete_graph(100)
nx.draw(H)
```

```python
I = nx.gnp_random_graph(20,0.5)
nx.draw(I)
```

```python
G = networkx.Graph()
G.add_node(1)
G.add_node(2)
G.add_node(3)
G.add_node(4)
G.add_node(5)
```
